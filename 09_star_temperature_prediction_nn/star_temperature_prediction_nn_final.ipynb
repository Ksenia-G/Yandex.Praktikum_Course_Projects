```python
# Function to display the main information about a data frame
def display_main_info(df):
    """
    Displays key information about the DataFrame for exploratory data analysis.
    
    Args:
        df (pandas.DataFrame): The DataFrame to be analyzed.
        
    Displays:
        - The first few rows of the DataFrame.
        - Data types and non-null counts of each column.
        - The shape (number of rows and columns) of the DataFrame.
        - The number of duplicate rows in the DataFrame.
        - The proportion of missing data for each column.
    """
    
    print('________________Head________________')
    display(df.head())
    
    print('________________Data Types Info________________')
    display(df.info())
    
    print('________________Shape________________')
    display(df.shape)
    
    print('________________Duplicates________________')
    display(df.duplicated().sum())
    
    print('________________Proportion of Missing Data________________')
    display(df.isna().mean())
```

```python
# Function to display statistical info
def show_stats(column_name: str):
    """
    Display summary statistics for a specified column in the dataset.

    """

    print(f'Statistics for {column_name}')
    display(df[[column_name]].describe().T)
```

```python
# Function to display density and box plots for numeric data
def plot_density_and_boxplot(data, column):
    """
    Plot density plot and boxplot for a given column in a DataFrame.

    Parameters:
    - df: pandas DataFrame
    - column: str

    Returns:
    - None
    """
    fig, axs = plt.subplots(1, 2, figsize=(12, 6))

    # Plot density plot
    sns.histplot(data[column], bins=50, color='lightgreen', kde=True, ax=axs[0])
    axs[0].set_title(f'Density Plot of {column}')

    # Plot boxplot
    axs[1].boxplot(data[column], vert=False, patch_artist=True, boxprops=dict(facecolor='lightblue'))
    axs[1].set_title(f'Boxplot of {column}')

    # Adjust layout
    plt.tight_layout()

    # Show plot
    plt.show()
```

```python
# Function that replaces rare categories based on a threshold
def remove_rare_categories(df, column_name, threshold):
    """
    Replaces rare categories in a specified column with 'Other' based on a threshold.

    Parameters:
    - df: pandas DataFrame
    - column_name: str, name of the column to process
    - threshold: float, categories with frequency <= threshold will be replaced

    Returns:
    - Updated Series with replaced rare categories and their normalized value counts.
    """

    category_counts = df[column_name].value_counts(normalize=True)
    rare_categories = category_counts[category_counts <= threshold].index
    df[column_name + '_new'] = df[column_name].apply(lambda x: 'Other' if x in rare_categories else x)
    
    return df[column_name + '_new'].value_counts(normalize=True)
```

```python
# Function that displays a bar chart for discrete values
def plot_bar_chart(data, column, dataset_name="Dataset"):

    value_counts = data[column].value_counts().sort_index()
    fig, ax = plt.subplots(figsize=(10, 6))
    ax.bar(value_counts.index, value_counts.values, color='yellow', edgecolor='black')

    ax.set_title(f"Bar Chart of {column}", fontsize=16)
    ax.set_xlabel(column, fontsize=14)
    ax.set_ylabel("Frequency", fontsize=14)
    ax.grid(axis='y', linestyle='--', alpha=0.7)

    plt.tight_layout()

    plt.show()
```

```python
# Function that plots a Phi_k correlation matrix
def plot_phik_heatmap(df, interval_cols, title='Phi_k Correlation Matrix'):
    """
    Plot a heatmap with a green color scheme for the Phi_k correlation matrix.

    Parameters:
    - df: pandas DataFrame
    - interval_cols: list of column names to be considered as interval columns for the Phi_k matrix
    - title: str, title of the plot (default is 'Phi_k Correlation Matrix')

    Returns:
    - None
    """
    # Calculate Phi_k matrix
    matrix = phik_matrix(df, interval_cols=interval_cols)
    
    # Plot heatmap
    plt.figure(figsize=(10, 8))
    sns.heatmap(matrix, annot=True, cmap='Greens', vmin=-1, vmax=1)
    plt.title(title)
    plt.show()
```

```python
# Input and output numbers of neurons
input_dim = X_train_tensor.shape[1]
output_dim = 1
```

```python
# Define a simple neural network class with 1 hidden layer
class NeuralNetwork(nn.Module):
    def __init__(self, input_dim, n_hidden_1, output_dim):
        super(NeuralNetwork, self).__init__()

        self.fc1 = nn.Linear(input_dim, n_hidden_1)
        self.act1 = nn.ReLU()

        self.fc2 = nn.Linear(n_hidden_1, output_dim)

    def forward(self, x):
        x = self.fc1(x)
        x = self.act1(x)

        x = self.fc2(x)
        return x
```

```python
# List of different values for the number of hidden neurons
n_hidden_neurons = [input_dim, 16, 32, 64]
n_epochs = 1000 

# Loss function (Mean Squared Error) for training
loss = nn.MSELoss()
```

```python
# Function to train the neural network
def train_neural_network(neural_network, n_epochs, n_layers=1):
    """
    This function trains the neural network model for a given number of epochs and calculates the RMSE on validation data.
    
    Parameters:
    neural_network (nn.Module): The neural network model to be trained.
    n_epochs (int): The number of epochs for training the model.
    n_layers (int, optional): The number of layers in the neural network. Default is 1.
    
    The function prints the RMSE value at the end of the training process for each epoch.
    """
    for epoch in range(n_epochs):
        neural_network.train()
        optimizer.zero_grad()
        
        y_pred = neural_network(X_train_tensor).flatten()
        loss_value = loss(y_pred, y_train_tensor)
        
        loss_value.backward() 
        
        optimizer.step()


    neural_network.eval()
    
    with torch.no_grad():
        test_preds = neural_network.forward(X_test_tensor)
        rmse = root_mean_squared_error(y_test_tensor, test_preds)
        
        if n_layers == 1:
            print(f"Neurons: {number}, RMSE: {rmse:.4f}")
        else:
            print(f"Neurons_1: {number_1}, Neurons_2: {number_2}, RMSE: {rmse:.4f}")
```

```python
# Loop over different numbers of hidden neurons and train the model for each
for number in n_hidden_neurons:

    baseline_1 = NeuralNetwork(input_dim=input_dim, n_hidden_1=number, output_dim=output_dim)

    optimizer = torch.optim.Adam(baseline_1.parameters(), lr=1e-3)
    
    train_neural_network(baseline_1, n_epochs)
```

## 3.3 –ë–∞–∑–æ–≤–∞—è –∫–æ–Ω—Ñ–∏–≥—É—Ä–∞—Ü–∏—è 2 (2 —Å–∫—Ä—ã—Ç—ã—Ö —Å–ª–æ—è)

```python
# Define a simple neural network class with 2 hidden layers
class NeuralNetwork(nn.Module):
    def __init__(self, input_dim, n_hidden_1, n_hidden_2, output_dim):
        super(NeuralNetwork, self).__init__()

        # First fully connected layer
        self.fc1 = nn.Linear(input_dim, n_hidden_1)
        self.act1 = nn.ReLU()

        # Second fully connected layer
        self.fc2 = nn.Linear(n_hidden_1, n_hidden_2)
        self.act2 = nn.ReLU()

        # Third fully connected layer
        self.fc3 = nn.Linear(n_hidden_2, output_dim)

    def forward(self, x):
        # Forward pass through layers and activations
        x = self.fc1(x)
        x = self.act1(x)

        x = self.fc2(x)
        x = self.act2(x)

        x = self.fc3(x)

        return x
```

```python
# List of different values for the number of hidden neurons
n_hidden_neurons_1 = [16, 32, 64]
n_hidden_neurons_2 = [16, 32, 64]
```

```python
# Loop over different combinations of hidden neurons and train the model for each
for number_1 in n_hidden_neurons_1:
    for number_2 in n_hidden_neurons_2:
        baseline_2 = NeuralNetwork(input_dim=input_dim, n_hidden_1=number_1, n_hidden_2=number_2, output_dim=output_dim)
        optimizer = torch.optim.Adam(baseline_2.parameters(), lr=0.1)
        train_neural_network(baseline_2, n_epochs, n_layers=2)
```

```python
# Set the parameters
n_hidden_1 = 16
n_hidden_2 = 16
dropout_rates = [0.0, 0.5, 0.9]
batch_sizes = [60, 120, 240]

# –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏
class NeuralNetwork(nn.Module):
    def __init__(self, input_dim, n_hidden_1, n_hidden_2, output_dim, dropout_rate=0.0):
        super(NeuralNetwork, self).__init__()
        self.fc1 = nn.Linear(input_dim, n_hidden_1)
        self.act1 = nn.ReLU()
        self.dropout1 = nn.Dropout(dropout_rate)
        self.fc2 = nn.Linear(n_hidden_1, n_hidden_2)
        self.act2 = nn.ReLU()
        self.dropout2 = nn.Dropout(dropout_rate)
        self.fc3 = nn.Linear(n_hidden_2, output_dim)

    def forward(self, x):
        x = self.fc1(x)
        x = self.act1(x)
        x = self.dropout1(x)
        x = self.fc2(x)
        x = self.act2(x)
        x = self.dropout2(x)
        x = self.fc3(x)
        return x

train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
test_dataset = TensorDataset(X_test_tensor, y_test_tensor)

# –¶–∏–∫–ª –æ–±—É—á–µ–Ω–∏—è –¥–ª—è —Ä–∞–∑–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π dropout_rate –∏ batch_size
for dropout_rate in dropout_rates:
    for batch_size in batch_sizes:
        print(f"Training with dropout_rate={dropout_rate}, batch_size={batch_size}")
        

        train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
        test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)
        
        model = NeuralNetwork(input_dim=input_dim, n_hidden_1=n_hidden_1, n_hidden_2=n_hidden_2, output_dim=output_dim, dropout_rate=dropout_rate)
        optimizer = torch.optim.Adam(model.parameters(), lr=0.1)

       
        for epoch in range(n_epochs):
            model.train()
            for X_batch, y_batch in train_loader:
                optimizer.zero_grad()
                y_pred = model(X_batch).flatten()
                loss_value = loss(y_pred, y_batch)
                loss_value.backward()
                optimizer.step()

        model.eval()
        with torch.no_grad():
            test_preds = []
            test_targets = []
            for X_batch, y_batch in test_loader:
                preds = model(X_batch).flatten()
                test_preds.append(preds)
                test_targets.append(y_batch)

            test_preds = torch.cat(test_preds)
            test_targets = torch.cat(test_targets)
            rmse = root_mean_squared_error(test_targets, test_preds)
            print(f"Dropout: {dropout_rate}, Batch Size: {batch_size}, RMSE: {rmse}")
```

- –ò—Å–ø–æ–ª—å–∑—É–µ–º —É–∂–µ –æ–±—É—á–µ–Ω–Ω—É—é –º–æ–¥–µ–ª—å —Å –ª—É—á—à–∏–º–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏ –ø–æ—Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫.

```python
dropout_rate_best = 0.5  # –õ—É—á—à–∏–π dropout
batch_size_best = 240  # –õ—É—á—à–∏–π batch size

n_hidden_1 = 16
n_hidden_2 = 16
input_dim = X_train_tensor.shape[1]  
output_dim = 1  
n_epochs = 50  
loss_fn = nn.MSELoss()  

train_loader = DataLoader(train_dataset, batch_size=batch_size_best, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=batch_size_best, shuffle=False)

model = NeuralNetwork(input_dim=input_dim, n_hidden_1=n_hidden_1, n_hidden_2=n_hidden_2, output_dim=output_dim, dropout_rate=dropout_rate_best)
optimizer = torch.optim.Adam(model.parameters(), lr=0.1)

for epoch in range(n_epochs):
    model.train()
    for X_batch, y_batch in train_loader:
        optimizer.zero_grad()
        y_pred = model(X_batch).flatten()
        loss_value = loss_fn(y_pred, y_batch)
        loss_value.backward()
        optimizer.step()

model.eval()
with torch.no_grad():
    test_preds = []
    test_targets = []
    for X_batch, y_batch in test_loader:
        preds = model(X_batch).flatten()
        test_preds.append(preds)
        test_targets.append(y_batch)

    test_preds = torch.cat(test_preds)
    test_targets = torch.cat(test_targets)


# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞ –¥–ª—è –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã—Ö –∏ —Ä–µ–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π
predicted = pd.DataFrame(test_preds.numpy(), columns=['temperature'])
actual = pd.DataFrame(test_targets.numpy(), columns=['temperature'])

fig, ax = plt.subplots(figsize=[20, 6])
indices = range(len(predicted))
plt.bar(indices, predicted['temperature'], color='steelblue', label='Predicted', alpha=0.7, width=0.4, align='center')
plt.bar(indices, actual['temperature'], color='coral', label='Actual', alpha=0.7, width=0.2, align='edge')

plt.title('Star Temperature (K) Predicted vs Actual', fontsize=16)
plt.xlabel('Stars', fontsize=14)
plt.ylabel('Temperature (K)', fontsize=14)
ax.legend()

plt.show()
```
